```python
print(df.columns)

y= df['Delivery Time']
x1= df['Sorting Time']
```

```python
plt.scatter(x1,y)

plt.xlabel("Sorting time", fontsize = 20)

plt.ylabel("Delivery time", fontsize = 20)

plt.show()
```

then i made the scatter diagram to plot all the data

```python
x = sm.add_constant(x1)
results=sm.OLS(y,x).fit()
results.summary()
```

i added a constant thats called the b0 in my regression model then i ran the OLS function on my data and i priinted it out to get the value of b1 and b0 to fit it in my regression line with equation yhat=b0+b1x1 i also concluded that they are strongly related since the value of p is zero which is optimal

```python
plt.scatter(x1,y)

plt.scatter(x1,y)

yhat = 1.6490*x1 + 6.5827

fig = plt.plot(x1,yhat, lw=4, c='blue', label = 'regression line')

plt.xlabel('Sorting time', fontsize = 20)

plt.ylabel('Delivery time', fontsize = 20)

plt.show()

```
